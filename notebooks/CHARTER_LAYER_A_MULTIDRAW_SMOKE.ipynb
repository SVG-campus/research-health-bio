# CHARTER_LAYER_A_MULTIDRAW_SMOKE

Hub-free **Layer A** primitive: synthetic two-sample mean gap with a **pooled label-shuffle permutation** null, budget capped by `runs/smoke.yaml` (`permutation_test.n_perm_max`). Runs in CI **before** the pillar domain stream charter row—complements streaming smokes, not domain truth.


```python
from methodology_preamble import init_notebook, build_run_card, assert_run_card
import numpy as np

info = init_notebook()
caps = info["methodology_caps"]["permutation_test"]
n_perm = max(39, min(149, int(caps["n_perm_max"])))

rng = np.random.default_rng(info["seed"] + 17)
a = rng.normal(0.0, 1.0, 50)
b = rng.normal(0.12, 1.0, 50)
obs = float(np.mean(a) - np.mean(b))


def _perm_stat(pool: np.ndarray, seed: int) -> float:
    r = np.random.default_rng(seed)
    idx = r.permutation(pool.size)
    sh = pool[idx]
    a2 = sh[:50]
    b2 = sh[50:]
    return float(np.mean(a2) - np.mean(b2))


pool = np.concatenate([a, b])
ge = 1
for i in range(n_perm):
    st = _perm_stat(pool, info["seed"] + 3000 + i)
    if abs(st) >= abs(obs):
        ge += 1
p = ge / (n_perm + 1)

rc = build_run_card(
    run_id="charter_layer_a_multidraw_smoke",
    hypothesis="Synthetic two-sample mean difference; pooled label-shuffle permutation null",
    metric=f"mean_diff_obs={obs:.4f} p_perm={p:.4f} n_perm={n_perm}",
    null="label_shuffle_within_pooled_vector",
    truth_scope="Synthetic RNG only; methodological sanity",
    seed=info["seed"],
)
assert_run_card(rc)
(obs, p, n_perm)
```
