# SMOKE_BOOTSTRAP
Percentile bootstrap mean CI using caps from `runs/smoke.yaml`.


```python
from methodology_preamble import init_notebook, build_run_card, assert_run_card
import numpy as np

info = init_notebook()
caps = info["methodology_caps"]["bootstrap_mean_ci"]
n_boot = min(400, int(caps["n_boot_max"]))
alpha = float(caps["alpha"])
rng = np.random.default_rng(info["seed"])
x = rng.normal(loc=0.5, scale=1.0, size=50)


def boot_mean(arr: np.ndarray, n: int, seed: int) -> tuple[float, float, float]:
    r = np.random.default_rng(seed)
    a = np.asarray(arr, dtype=float).ravel()
    boots = r.choice(a, size=(n, a.size), replace=True).mean(axis=1)
    lo_q = (alpha / 2) * 100
    hi_q = (1 - alpha / 2) * 100
    return float(a.mean()), float(np.percentile(boots, lo_q)), float(np.percentile(boots, hi_q))


m, lo, hi = boot_mean(x, n_boot, info["seed"] + 3)
rc = build_run_card(
    run_id="nb_smoke_bootstrap",
    hypothesis="Bootstrap smoke: mean CI finite on synthetic normal",
    metric=f"mean={m:.4f} ci=({lo:.4f},{hi:.4f}) n_boot={n_boot}",
    null="n/a (bootstrap is uncertainty, not a null test)",
    truth_scope="Synthetic only",
    seed=info["seed"],
)
assert_run_card(rc)
(m, lo, hi)
```
