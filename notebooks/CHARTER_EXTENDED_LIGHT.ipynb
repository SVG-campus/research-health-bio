# CHARTER_EXTENDED_LIGHT
Light **Layer A-shaped** step: one synthetic draw + mean + run card (not domain truth).
Use as the bridge between `CHARTER_SHELL` and a full charter notebook.


```python
from methodology_preamble import init_notebook, build_run_card, assert_run_card
import numpy as np

info = init_notebook()
rng = np.random.default_rng(info["seed"] + 11)
x = rng.normal(loc=0.0, scale=1.0, size=40)
m = float(x.mean())
rc = build_run_card(
    run_id="charter_extended_light_smoke",
    hypothesis="Synthetic mean near 0 (RNG sanity)",
    metric=f"mean_n40={m:.4f}",
    null="n/a (descriptive draw only)",
    truth_scope="Synthetic RNG only",
    seed=info["seed"],
)
assert_run_card(rc)
m
```
